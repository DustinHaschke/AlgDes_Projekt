```python
import matplotlib.pyplot as plt

reisezeiten_namen = ["Filmhaus", "InfinityDome", "Kolosseum", "Koki", "CineStar"]
reisezeiten = [
    [0, 3, 6, 2, 4],  # Von Filmhaus zu ...
    [3, 0, 4, 2, 1],  # Von InfinityDome ...
    [7, 4, 0, 8, 2],  # Von Kolosseum ...
    [4, 3, 7, 0, 4],  # Von Koki ...
    [4, 2, 2, 5, 0]   # Von CineStar ...
]

zulaessige_ueberschneidungsminuten = 1

#ermittelt die Reisezeiten vom aktuellen Startpunkt aus
def ort_zu_reisezeit(ort):
    for i, name in enumerate(reisezeiten_namen):
        if name.lower() == ort.split(" ")[0].lower():
            return i
    print("Fehler Unbekannter Ort:", ort)

# gibt die Gesamtzeit für den Tag zurück
def startzeit_ermittlung(hhmm):
    hh, mm = hhmm.split(":")
    return int(hh) * 60 + int(mm)


def datenvorverarbeitung(line):
    parts = line.split(";")
    full_title = parts[0].strip()
    wochentag = parts[1].strip()
    startzeit = startzeit_ermittlung(parts[2].strip())
    dauer = int(parts[3].strip())
    ort_name = parts[4].strip()
    bewertung = float(parts[5].strip())

    return {
        "kurztitel": full_title.split("|")[0].strip(),
        "filmtitel": full_title,
        "startzeit": startzeit,
        "endzeit": startzeit + dauer,
        "dauer": dauer,
        "orts_index": ort_zu_reisezeit(ort_name),
        "ort_string": ort_name,
        "bewertung": bewertung,
        "wochentag": wochentag
    }

# Prüft, ob die Funktionerreichbar ist
def ist_erreichbar(ei, ej):
    reisezeit = reisezeiten[ei["orts_index"]][ej["orts_index"]]
    return ej["startzeit"] >= ei["endzeit"] + reisezeit - zulaessige_ueberschneidungsminuten

def gewichtetes_interval_scheduling(events):
    #sortieren der Events nach dem Ende der Vorstellung
    events.sort(key=lambda x: x["endzeit"])
    p = [0] * len(events)

    #durchlaufe alle Events und prüfe, ob Knoten erreichbar sind
    for j in range(len(events)):
        ej = events[j]
        i_found = -1
        for i in range(j - 1, -1, -1):
            if ist_erreichbar(events[i], ej):
                i_found = i
                break
        p[j] = i_found

    dp = [0] * (len(events) + 1)

    #bewerte_gefundene Pfade
    for j in range(1, len(events) + 1):
        event_index = j - 1
        bewertung_j = events[event_index]["bewertung"]
        i = p[event_index]
        wert_wenn_belegt = bewertung_j + (dp[i + 1] if i != -1 else 0)
        dp[j] = max(dp[j - 1], wert_wenn_belegt)

    gewaehlte_indizes = []
    j = len(events)
    while j > 0:
        if dp[j] == dp[j - 1]:
            j -= 1
        else:
            event_index = j - 1
            gewaehlte_indizes.append(event_index)
            i = p[event_index]
            j = i + 1 if i != -1 else 0

    gewaehlte_indizes.reverse()
    return dp[len(events)], [events[idx] for idx in gewaehlte_indizes]



# Plotten einer lesbaren Timetable
def timetable_plotten(datensatz_name, events, gewaehlte_filme):
    einzigartige_filmliste = list({event["kurztitel"] for event in gewaehlte_filme})
    farben = {film: plt.cm.tab20(i % 20) for i, film in enumerate(einzigartige_filmliste)}

    wochentags_order = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]

    fig, ax = plt.subplots(figsize=(18, 8))
    label_liste = []

    for event in gewaehlte_filme:
        startzeit = event["startzeit"]
        endzeit = event["endzeit"]
        wochentag = event["wochentag"]
        color = farben[event["kurztitel"]]
        x_pos = wochentags_order.index(wochentag) + 0.5

        ax.broken_barh(
            [(x_pos - 0.4, 0.8)],
            (startzeit / 60, (endzeit - startzeit) / 60),
            facecolors=color,
            edgecolor="black",
        )

        legend_label = f"{event['kurztitel']} ({wochentag.capitalize()}, {startzeit // 60:02d}:{startzeit % 60:02d} - {endzeit // 60:02d}:{endzeit % 60:02d})"
        if legend_label not in label_liste:
            label_liste.append(legend_label)
            ax.bar(0, 0, color=color, label=legend_label)

    ax.set_xticks([i + 0.5 for i in range(len(wochentags_order) + 1)])
    ax.set_xticklabels([wochentag.capitalize() for wochentag in wochentags_order] + ["Next"])
    ax.set_yticks(range(0, 25, 3))
    ax.set_yticklabels([f"{hour}:00" for hour in range(0, 25, 3)])
    ax.set_ylabel("Stunden")
    ax.set_xlabel("Wochentag")
    ax.set_title(f"Optimale Timetable: {datensatz_name}")
    ax.grid(axis="y", linestyle="--", alpha=0.7)
    ax.grid(axis="x", linestyle="--", alpha=0.7)
    ax.set_xlim(0, len(wochentags_order))
    ax.set_ylim(0, 24)
    ax.legend(
        loc="upper left",
        bbox_to_anchor=(1.01, 1),
        title="Filme",
        fontsize="small",
    )
    plt.tight_layout()
    plt.show()

def verarbeite_datensets(datendict):
    for datensatz_name, dataset in datendict.items():
        all_events = [datenvorverarbeitung(line) for line in dataset.split("\n")]

        best_event_by_short = {}
        for e in all_events:
            sn = e["kurztitel"]
            if sn not in best_event_by_short or e["bewertung"] > best_event_by_short[sn]["bewertung"]:
                best_event_by_short[sn] = e

        events = list(best_event_by_short.values())
        best_total_bewertung, gewaehlte_filme = gewichtetes_interval_scheduling(events)

        print(f"Datensatz: {datensatz_name}")
        # gibt eine gerundete Gesamtbewertung aus und prüft auf übereinstimmenden angegebenen Wert
        print("Maximale Gesamtbewertung =", round(best_total_bewertung, 4))
        if abs(round(best_total_bewertung, 4) - erwartete_ergebnisse[datensatz_name]) <0.1:
            print("Ergebnis stimmt mit dem angegebenen Erwartungswert überein.")
        else:
            print("Ergebnis stimmt nicht mit dem angegebenen Erwartungswert überein❗")

        print("optimaler Terminplan:")
        for evt in gewaehlte_filme:
            s_h = evt["startzeit"] // 60
            s_m = evt["startzeit"] % 60
            e_h = evt["endzeit"] // 60
            e_m = evt["endzeit"] % 60
            print(f" - {evt['kurztitel']} ({evt['wochentag']}): "
                  f"{s_h:02d}:{s_m:02d} - {e_h:02d}:{e_m:02d} "
                  f"(Ort: {evt['ort_string']}, bewertung={evt['bewertung']})")

        timetable_plotten(datensatz_name, events, gewaehlte_filme)

# small-Datensätze
gesamtdatensatz_small = {
    "small101": """Omnejd;wed;19:00;42;InfinityDome;7.1
VR Expanded;wed;20:30;84;InfinityDome;7.7
Dancing Queen;thu;10:15;90;CineStar 1;7.5
VR Expanded;thu;17:00;84;InfinityDome;7.7
Omnejd;thu;18:45;42;InfinityDome;7.1
Overview;fri;11:15;32;InfinityDome;8.7
Opus 136;fri;11:47;10;InfinityDome;6.3
VR Expanded;fri;18:15;84;InfinityDome;7.7
Non-Face;fri;21:00;45;InfinityDome;6.5
Shorts 4+: Spiel Kino;sat;10:00;14;Koki;7.7
VR Expanded;sat;16:15;84;InfinityDome;7.7
Sigade revolutsioon | Revolution of Pigs | Revolution der Schweine;sat;22:15;98;CineStar 4;6.6
Opus 136;sun;11:44;10;InfinityDome;6.3
Omnejd;sun;14:30;43;InfinityDome;7.1
VR Expanded;sun;18:15;84;InfinityDome;7.7""",

    "small839": """Titina | Titina – Ein tierisches Abenteuer am Nordpol;wed;16:15;92;CineStar 1;7.1
And the King Said, What a Fantastic Machine | Fantastic Machine;wed;16:45;85;CineStar 6;7.2
Nordic Shorts 1: Cracks in the Exterior;wed;16:45;71;CineStar 2;7.2
Elaha;wed;18:15;110;CineStar 7;7.6
Huset | Prisoner;wed;19:00;118;CineStar 5;7.6
Omnejd;wed;19:00;42;InfinityDome;7.1
Norwegian Dream;wed;19:15;97;CineStar 1;6.6
Ung Flukt | The Wayward Girl | Die jungen Sünder;wed;19:15;94;CineStar 4;6.6
Tillsammans 99 | Together 99;wed;19:30;115;Kolosseum;6.5
Nordic Shorts 2: Inbetween;wed;19:45;83;CineStar 2;8.1
Twice Colonized;wed;19:45;92;CineStar 6;6.4
Heima er best | Descendants;wed;20:00;90;Koki;5.9
VR Expanded;wed;20:30;84;InfinityDome;7.7
El sueño de la Sultana | Sultana's Dream | Sultanas Traum;wed;21:15;86;CineStar 7;6.8
Ilveskuiskaaja | Lynx Man;wed;22:00;82;CineStar 5;7.8
Nordic Shorts 3: Real Fake Truths;wed;22:15;74;CineStar 1;7.0
Poetas | The Poet;wed;22:15;112;CineStar 4;7.3
Revir – En fortælling om det dyrebare | Revir – Everything You Hold Dear;wed;22:30;83;Koki;6.8
Tilverur | Solitude;wed;22:30;75;Kolosseum;7.7
Die Saat | The Seed;wed;22:45;82;CineStar 6;7.3
Sånger från andra våningen | Songs from the Second Floor;wed;22:45;99;CineStar 2;7.5""",

    "small131":"""Elaha;wed;18:15;90;CineStar 7;7.6
Twice Colonized;wed;19:45;92;CineStar 6;6.4
Superposition;thu;13:30;102;CineStar 3;6.3
Neona Pavasaris | Neon Spring;thu;16:15;102;CineStar 1;5.9
Und dass man ohne Täuschung zu leben vermag | Of Living Without Illusion;thu;19:15;109;CineStar 7;5.8
Miraklet i Gullspång | The Gullspång Miracle;fri;10:45;109;CineStar 6;8.3
Om det oändliga | About Endlessness | Über die Unendlichkeit;fri;10:45;78;CineStar 2;6.8
Opus 136;fri;11:47;10;InfinityDome;6.3
Roselil og stentrolden | Rosa and the Stone Troll;fri;13:15;75;CineStar 1;7.3
Shorts 10+: Verborgene Orte;fri;16:15;78;CineStar 1;5.6
100 Jahre Planetarien;fri;16:45;58;InfinityDome;7.5
Paradiset Brinner | Paradise Is Burning;fri;19:30;108;CineStar 3;6.8
Ellos eatnu – La elva leve | Let the River Flow;fri;22:30;122;CineStar 3;7.7
Roselil og stentrolden | Rosa and the Stone Troll;sat;10:15;75;CineStar 1;7.3
Helke Sander: Aufräumen | Helke Sander: Cleaning House;sat;13:15;82;CineStar 7;7.7
Hugo Hercules & The Wild West / The Artist W. H. D. Koerner;sat;16:15;88;CineStar 7;7.1
Nelly Rapp – Dödens spegel | Nelly Rapp – The Secret of the Black Forest;sun;10:15;95;CineStar 1;6.7
And the King Said, What a Fantastic Machine | Fantastic Machine;sun;15:00;85;Koki;7.2
Tilverur | Solitude;sun;16:00;75;CineStar 5;7.7""",

    "small167":"""Future North Präsentation: Baltisches Filmschaffen;fri;10:45;45;Filmhaus 1;7.7
Shorts 10+: Verborgene Orte;fri;16:15;78;CineStar 1;5.6
Je'vida;fri;16:30;99;CineStar 3;7.5
Paradiset Brinner | Paradise Is Burning;fri;19:30;108;CineStar 3;6.8
Togoland Projections | Togoland Projektionen;sat;10:15;96;CineStar 7;5.7
Slow;sat;10:30;108;Kolosseum;7.5
Du levande | You, the Living | Das jüngste Gewitter;sat;10:45;93;CineStar 2;7.4
Terrorister – en film om dom dömda | Terrorists: The Kids They Sentenced;sat;13:15;85;CineStar 4;6.3
Fedrelandet | Songs of Earth;sat;13:45;90;CineStar 6;7.3
Vi lover et helvete | Saving the Fucking Planet E1-3;sat;16:45;210;CineStar 2;7.0
100 Jahre Planetarien;sat;18:00;58;InfinityDome;7.5
Nordic Shorts 4: The Nuclear Family;sun;10:00;73;Koki;6.5
Verden er uskarp | The World Is Out Of Focus;sun;10:30;71;Filmhaus 2;6.9
Å Øve | Practice;sun;13:00;79;CineStar 5;6.2
Ellos eatnu – La elva leve | Let the River Flow;sun;16:45;122;CineStar 2;7.7""",

    "small251":"""Omnejd;wed;19:00;42;InfinityDome;7.1
Waldhusener Forst – Wald voller Geschichten | Waldhusener Forst – Forest of Stories;thu;12:30;25;InfinityDome;6.4
Omnejd;thu;18:45;42;InfinityDome;7.1
Omnejd;thu;18:45;42;InfinityDome;7.1
100 Jahre Planetarien;thu;20:15;58;InfinityDome;7.5
Future North Präsentation: Baltisches Filmschaffen;fri;10:45;45;Filmhaus 1;7.7
Overview;fri;11:15;32;InfinityDome;8.7
Opus 136;fri;11:47;10;InfinityDome;6.3
Drones of the Deep | Drones of the Deep - multimedia presentation by Tom Kwasnitschka | Ein Spaziergang auf dem Meeresgrund – Multimedialer Vortrag von Tom Kwasnitschka;fri;15:15;45;InfinityDome;6.0
Filmforum Shorts: (Be)Rauschend;fri;16:15;54;CineStar 7;6.1
100 Jahre Planetarien;fri;16:45;58;InfinityDome;7.5
Non-Face;fri;21:00;45;InfinityDome;6.5
Shorts 4+: Spiel Kino;sat;10:00;14;Koki;7.7
Shorts 4+: Spiel Kino;sat;10:00;14;Koki;7.7
Heartist;sat;13:00;51;CineStar 5;7.9
Legend of the Enchanted Reef | Die Legendzeit des Zauberriffs;sat;14:00;43;InfinityDome;4.3
Overview;sat;15:15;32;InfinityDome;8.7
100 Jahre Planetarien;sat;18:00;58;InfinityDome;7.5
100 Jahre Planetarien;sat;18:00;58;InfinityDome;7.5
Um Menino;sun;10:00;46;InfinityDome;3.8
Heartist;sun;10:45;51;CineStar 6;7.9
Aurora - Wunderbares Polarlicht;sun;11:15;29;InfinityDome;5.7
Opus 136;sun;11:44;10;InfinityDome;6.3
Tauche mit uns zum Hanseschiff | Dive into the Hanse Shipwreck;sun;13:00;30;InfinityDome;8.1
Omnejd;sun;14:30;43;InfinityDome;7.1""",

    "small353":"""Heima er best | Descendants;wed;20:00;90;Koki;5.9
Revir – En fortælling om det dyrebare | Revir – Everything You Hold Dear;wed;22:30;83;Koki;6.8
Huset | Prisoner;thu;10:00;118;Koki;7.6
Nordic Shorts 2: Inbetween;thu;12:30;83;Koki;8.1
Det är en dag imorgon också | Tomorrow's Another wochentag;thu;15:00;90;Koki;7.4
Being a Human Person;thu;17:30;90;Koki;7.6
Estonia;thu;20:00;90;Koki;6.9
Miraklet i Gullspång | The Gullspång Miracle;thu;22:30;109;Koki;8.3
Päikeseaeg | Sundial;fri;10:00;65;Koki;6.6
Nordic Shorts 4: The Nuclear Family;fri;12:30;73;Koki;6.5
Handen på Hjärtat | Out of Touch;fri;15:00;132;Koki;6.9
Arkitekten | The Architect;fri;17:30;76;Koki;6.9
Heima er best | Descendants;fri;20:00;90;Koki;5.9
Svārstības | Before the Light;fri;22:30;75;Koki;7.4
Arkitekten | The Architect;sat;10:00;76;CineStar 5;6.9
Shorts 4+: Spiel Kino;sat;10:00;14;Koki;7.7
Shorts 4+: Spiel Kino;sat;10:00;14;Koki;7.7
Togoland Projections | Togoland Projektionen;sat;10:15;96;CineStar 7;5.7
Nordic Shorts 5: Broken Connections;sat;12:30;87;Koki;8.2
Shorts: Miteinander;sat;14:30;79;Koki;6.8
Shorts 16+: Wohin mit unseren Träumen?;sat;17:30;73;Koki;8.4
Nordic Shorts 1: Cracks in the Exterior;sat;20:00;71;Koki;7.2
Nordic Shorts 3: Real Fake Truths;sat;22:30;74;Koki;7.0
Nordic Shorts 4: The Nuclear Family;sun;10:00;73;Koki;6.5
Afturelding | Coach | Balls;sun;13:30;88;Koki;7.9
And the King Said, What a Fantastic Machine | Fantastic Machine;sun;15:00;85;Koki;7.2
Estonia;sun;17:30;90;Koki;6.9
Tillsammans! | Together! | Zusammen!;sun;20:00;106;Koki;7.4""",

    "small373":"""Bastarden | The Promised Land | King’s Land;thu;19:30;127;CineStar 3;7.8
Ellos eatnu – La elva leve | Let the River Flow;thu;19:30;122;Kolosseum;7.7
Handen på Hjärtat | Out of Touch;fri;15:00;132;Koki;6.9
Bastarden | The Promised Land | King’s Land;fri;22:30;127;Kolosseum;7.6
Ellos eatnu – La elva leve | Let the River Flow;fri;22:30;122;CineStar 3;7.7
Ellos eatnu – La elva leve | Let the River Flow;sat;10:30;122;CineStar 3;7.7
Bastarden | The Promised Land | King’s Land;sat;16:30;127;Kolosseum;7.6
Vi lover et helvete | Saving the Fucking Planet E1-3;sat;16:45;210;CineStar 2;7.0
Vi lover et helvete | Saving the Fucking Planet E4-6;sat;19:45;210;CineStar 2;6.8
Handen på Hjärtat | Out of Touch;sat;22:45;132;CineStar 2;6.9
Bastarden | The Promised Land | King’s Land;sun;13:30;127;CineStar 3;7.6
Ellos eatnu – La elva leve | Let the River Flow;sun;16:45;122;CineStar 2;7.7""",

    "small421":"""And the King Said, What a Fantastic Machine | Fantastic Machine;wed;16:45;85;CineStar 6;7.2
Für Immer | For Ever;thu;16:15;87;CineStar 7;7.5
Å Øve | Practice;thu;16:30;79;Kolosseum;6.2
Shorts 16+: Wohin mit unseren Träumen?;thu;19:45;73;CineStar 2;8.4
Slow;thu;22:00;108;CineStar 5;7.5
Å Øve | Practice;fri;10:30;79;CineStar 3;6.2
Om det oändliga | About Endlessness | Über die Unendlichkeit;fri;10:45;78;CineStar 2;6.8
Und dass man ohne Täuschung zu leben vermag | Of Living Without Illusion;fri;13:00;109;CineStar 5;5.8
Slow;fri;13:30;108;CineStar 3;7.5
Fedrelandet | Songs of Earth;fri;19:45;90;CineStar 6;7.3
Mein Falke | My Falcon;fri;22:15;105;CineStar 7;6.4
Slow;sat;10:30;108;Kolosseum;7.5
Å Øve | Practice;sat;13:30;79;Kolosseum;6.2
Fedrelandet | Songs of Earth;sat;13:45;90;CineStar 6;7.3
Shorts 16+: Wohin mit unseren Träumen?;sat;17:30;73;Koki;8.4
Morden im Norden – Am Abgrund;sat;22:15;88;CineStar 7;5.8
Tatort – Borowski und das unschuldige Kind von Wacken | Tatort – Borowski and the Innocent Child of Wacken;sun;10:15;88;CineStar 7;5.4
Å Øve | Practice;sun;13:00;79;CineStar 5;6.2
Piratene | Pirates | Die Piraten;sun;13:15;93;CineStar 4;5.9
And the King Said, What a Fantastic Machine | Fantastic Machine;sun;15:00;85;Koki;7.2
Slow;sun;19:30;108;Kolosseum;7.5""",

    "small449":"""Tillsammans 99 | Together 99;wed;19:30;115;Kolosseum;6.5
Für Immer | For Ever;thu;16:15;87;CineStar 7;7.5
Neona Pavasaris | Neon Spring;thu;16:15;102;CineStar 1;5.9
Vintersaga;thu;16:30;81;CineStar 3;7.3
Slow;thu;22:00;108;CineStar 5;7.5
Opus 136;fri;11:47;10;InfinityDome;6.3
Stille Liv | The Quiet Migration;fri;19:30;102;Kolosseum;6.3
Bastarden | The Promised Land | King’s Land;fri;22:30;127;Kolosseum;7.6
København findes ikke | Copenhagen Does Not Exist;sat;16:00;99;CineStar 5;5.9
Vintersaga;sun;10:30;81;CineStar 3;7.3
Tilverur | Solitude;sun;16:00;75;CineStar 5;7.7
Superposition;sun;16:30;102;Kolosseum;6.3""",

    "small643":"""Ung Flukt | The Wayward Girl | Die jungen Sünder;wed;19:15;94;CineStar 4;6.6
Roy Andersson: Die frühen Jahre;thu;10:45;71;CineStar 2;6.2
Gategutter | Boys From the Streets | Gassenjungen;thu;13:15;77;CineStar 4;6.5
The Small Confession + 511 Best Photographs of Mars;thu;16:15;98;CineStar 4;7.8
Flickorna Gyurkovics | A Sister of Six | Die sieben Töchter der Frau Gyurkovics;thu;19:15;102;CineStar 4;7.1
Käpy selän alla | Under Your Skin | Tannenzapfen unter dem Rücken;thu;22:15;89;CineStar 4;6.7
Opus 136;fri;11:47;10;InfinityDome;6.3
Dom kallar oss mods | They Call Us Misfits | Sie nennen uns Mods;fri;13:15;100;CineStar 4;7.3
A Sentimental Short Story + Motorcycle Summer;fri;16:15;81;CineStar 4;6.6
Arkitekten | The Architect;fri;17:40;76;Koki;6.9
Medan staden sover | While the City Sleeps;fri;19:15;102;CineStar 4;5.9
Lasse og Geir | Them and Us | Lasse und Geir;fri;22:15;91;CineStar 4;6.4
Spiele für Schulkinder + estnische Kurzfilme;sat;16:15;108;CineStar 4;6.6
Täältä tullaan, elämä! | Right On, Man! | Ich pfeif auf eure Hilfe!;sat;19:15;111;CineStar 4;6.8
Ung Flukt | The Wayward Girl | Die jungen Sünder;sun;10:15;94;CineStar 4;6.6
Opus 136;sun;11:44;10;InfinityDome;6.3
Piratene | Pirates | Die Piraten;sun;13:15;93;CineStar 4;5.9
Liv | Das Mädchen Liv;sun;16:15;88;CineStar 4;5.7""",

    "small743":"""Nordic Shorts 1: Cracks in the Exterior;wed;17:05;71;CineStar 2;8.6
Elaha;wed;18:15;110;CineStar 7;7.6
!Aitsa;thu;16:45;89;CineStar 6;8.8
Inger & det dårlige selskab | Mrs Hansen & The Bad Companions;thu;19:45;90;CineStar 6;8.2
Miraklet i Gullspång | The Gullspång Miracle;thu;22:30;109;Koki;8.3
Den siste filmen | The Last Movie;thu;22:45;83;CineStar 6;8.3
Miraklet i Gullspång | The Gullspång Miracle;fri;10:45;109;CineStar 6;8.3
Cinéma Laika | Cinema Laika | Kino Laika;fri;13:45;81;CineStar 6;8.1
Den siste filmen | The Last Movie;sat;10:15;83;CineStar 4;8.3
Inger & det dårlige selskab | Mrs Hansen & The Bad Companions;sat;10:30;90;Filmhaus 2;8.2
Cinéma Laika | Cinema Laika | Kino Laika;sat;22:45;81;CineStar 6;8.1
!Aitsa;sun;19:45;89;CineStar 6;8.8""",

    "small751":"""Titina | Titina – Ein tierisches Abenteuer am Nordpol;wed;16:15;92;CineStar 1;7.1
Norwegian Dream;wed;19:15;97;CineStar 1;6.6
Dancing Queen;thu;10:15;90;CineStar 1;7.5
Nelly Rapp – Dödens spegel | Nelly Rapp – The Secret of the Black Forest;thu;13:15;95;CineStar 1;6.7
Blomster;thu;13:45;65;CineStar 2;5.9
Neona Pavasaris | Neon Spring;thu;16:15;102;CineStar 1;5.9
Shorts 14+: Eine Handbreit über dem Boden;thu;16:45;73;CineStar 2;7.3
De umyndige | Bravehearts;thu;19:15;71;CineStar 1;7.1
Shorts 16+: Wohin mit unseren Träumen?;thu;19:45;73;CineStar 2;8.4
Forvandlingen – frihed kommer indefra | It's Always Been Me;fri;10:15;72;CineStar 1;7.0
Roselil og stentrolden | Rosa and the Stone Troll;fri;13:15;75;CineStar 1;7.3
Hør her'a! | Listen Up!;fri;13:45;86;CineStar 2;6.9
Shorts 10+: Verborgene Orte;fri;16:15;78;CineStar 1;5.6
Norwegian Dream;fri;19:15;97;CineStar 1;6.6
Shorts 4+: Spiel Kino;sat;10:00;14;Koki;7.7
Roselil og stentrolden | Rosa and the Stone Troll;sat;10:15;75;CineStar 1;7.3
Shorts 10+: Verborgene Orte;sat;13:15;78;CineStar 1;5.6
Neona Pavasaris | Neon Spring;sat;13:45;102;CineStar 2;5.9
Titina | Titina – Ein tierisches Abenteuer am Nordpol;sat;16:15;92;CineStar 1;7.1
Shorts 16+: Wohin mit unseren Träumen?;sat;17:30;73;Koki;8.4
De umyndige | Bravehearts;sat;19:15;71;CineStar 1;7.1
Nelly Rapp – Dödens spegel | Nelly Rapp – The Secret of the Black Forest;sun;10:15;95;CineStar 1;6.7
Hør her'a! | Listen Up!;sun;10:45;86;CineStar 2;6.9
Dancing Queen;sun;13:15;90;CineStar 1;7.5
Blomster;sun;13:45;65;CineStar 2;5.9
Forvandlingen – frihed kommer indefra | It's Always Been Me;sun;16:15;72;CineStar 1;7.0
Shorts 14+: Eine Handbreit über dem Boden;sun;19:15;73;CineStar 1;7.3""",

    "small797":"""København findes ikke | Copenhagen Does Not Exist;thu;10:30;99;CineStar 3;5.9
Blomster;thu;13:45;65;CineStar 2;5.9
Neona Pavasaris | Neon Spring;thu;16:15;102;CineStar 1;5.9
Und dass man ohne Täuschung zu leben vermag | Of Living Without Illusion;thu;19:15;109;CineStar 7;5.8
København findes ikke | Copenhagen Does Not Exist;fri;10:15;99;CineStar 4;5.9
Und dass man ohne Täuschung zu leben vermag | Of Living Without Illusion;fri;13:00;109;CineStar 5;5.8
Medan staden sover | While the City Sleeps;fri;19:15;102;CineStar 4;5.9
Neona Pavasaris | Neon Spring;sat;13:45;102;CineStar 2;5.9
Legend of the Enchanted Reef | Die Legendzeit des Zauberriffs;sat;14:00;43;InfinityDome;4.3
København findes ikke | Copenhagen Does Not Exist;sat;16:00;99;CineStar 5;5.9
Morden im Norden – Am Abgrund;sat;22:15;88;CineStar 7;5.8
Um Menino;sun;10:00;49;InfinityDome;3.8
Blomster;sun;13:45;65;CineStar 2;5.9
Liv | Das Mädchen Liv;sun;16:15;88;CineStar 4;5.7
København findes ikke | Copenhagen Does Not Exist;sun;19:00;99;CineStar 5;5.9"""


}

erwartete_ergebnisse = {
    "small101": 58.1,
    "small131" : 98.8,
    "small167" : 65.1,
    "small251" : 99.8,
    "small353": 160.7,
    "small373" : 29.4,
    "small421" : 80.2,
    "small449" : 62.6,
    "small643" : 105.3,
    "small743" : 57.9,
    "small751" : 97.0,
    "small797" : 49.0,
    "small839" : 36.7
}

# Verarbeitung aller Datensätze
verarbeite_datensets(gesamtdatensatz_small)
```
